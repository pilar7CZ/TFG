```python
import sys
import os

import supervision as sv

from PIL import Image
from ultralytics import YOLO
```

```python
#CREATE A NEW YOLO MODEL
model_sc = YOLO("yolov8n.yaml")
```

```python
results_sc = model_sc.train(data="coco128.yaml", epochs=3)
```

```python
img_personas1 = Image.open(ruta_personas)
res_personas1 = model_sc.predict(source=img_personas1, save=True, save_txt=True)

img_bus1 = Image.open(ruta_bus)
res_bus1 = model_sc.predict(source=img_bus1, save=True, save_txt=True)

img_coche1 = Image.open(ruta_coche)
res_coche1 = model_sc.predict(source=img_coche1, save=True, save_txt=True)

img_coche2 = Image.open(ruta_coche2)
res_coche2 = model_sc.predict(source=img_coche2, save=True, save_txt=True)
```
