```python
import os
import json
from google.colab import drive
drive.mount('/content/drive')

def convert_to_yolo(json_file, output_folder):

    # Abrimos el archivo en modo lectura
    with open(json_file, 'r') as f:
        data = json.load(f)

    img_width = data['imageWidth']
    img_height = data['imageHeight']

    # Crear el archivo de salida
    output_file = os.path.join(output_folder, os.path.splitext(os.path.basename(json_file))[0] + '.txt')

    with open(output_file, 'w') as f:
        for shape in data['shapes']:
            class_id = shape['group_id']
            x_min = min(point[0] for point in shape['points'])
            y_min = min(point[1] for point in shape['points'])
            x_max = max(point[0] for point in shape['points'])
            y_max = max(point[1] for point in shape['points'])

            # Convertir coordenadas a relativas
            x_center = (x_min + x_max) / (2 * img_width)
            y_center = (y_min + y_max) / (2 * img_height)
            width = (x_max - x_min) / img_width
            height = (y_max - y_min) / img_height

            # Escribir en el archivo de salida
            f.write(f"{class_id} {x_center} {y_center} {width} {height}\n")

# USO DEL SCRIPT

#Ruta de la carpeta que quiero recorrer
folder = "/content/drive/My Drive/Colab Notebooks/coches"
output_folder = "/content/drive/My Drive/Colab Notebooks/coches/dataset_example"

# Recorro todos los archivos de la carpeta
for nombre_archivo in os.listdir(folder):

  # Concateno la ruta de la carpeta con el nombre del archivo
  file_route = os.path.join(folder, nombre_archivo)


  #Verifica si 'file_route' es un archivo (no una carpeta)
  if os.path.isfile(file_route):

    # Verifico si el archivo es un json
    if nombre_archivo.endswith('.json'):
      convert_to_yolo(file_route, output_folder)
      print(file_route)


```
